```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_name(filePath, run):
    fileName = f"{filePath}/X03MA_PC_{run:04d}"
    return fileName
```

```python
filePath = "C:/Researches/Data/VI3/202211/MDA/"

def load_scan(filePath,run):
    fileName = make_name(filePath,run)
    d1 = np.loadtxt(f"{fileName}_D0.txt",comments='#',delimiter=',')
    d2 = np.loadtxt(f"{fileName}_D1.txt",comments='#',delimiter=',')
    Y = np.loadtxt(f"{fileName}_P0.txt",comments='#',delimiter=',')
    Z = np.loadtxt(f"{fileName}_P1.txt",comments='#',delimiter=',')

    return Y,Z,d1,d2

def plot_scan(X,Y,data):
    fig, ax = plt.subplots()
    fig.gca().set_aspect(1)
    im = ax.pcolor(X,Y,data,shading='nearest',cmap='jet')
    ax.invert_yaxis()
    ax.set_xlabel("y ( mm )")
    ax.set_ylabel("z ( mm )")
    fig.colorbar(im)

    return fig,ax,im
```

```python
Y,Z,data1,d2 = load_scan(filePath,7)
plot_scan(Y,Z,data1)
Y,Z,data2,d2 = load_scan(filePath,8)
plot_scan(Y,Z,data2)
Y,Z,data3,d2 = load_scan(filePath,9)
plot_scan(Y,Z,data3)
Y,Z,data4,d2 = load_scan(filePath,10)
plot_scan(Y,Z,data4)
```

```python
data = - (data1-data4) + (data2-data3)
fig,ax,im = plot_scan(Y,Z,data)

im.set_clim(0,0.03)
im.set_cmap("jet")
```
